```python
import pandas as pd

# Load the data
data = pd.read_csv("census.csv")

# Display the first few rows to inspect
data.head()

```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
categorical_columns = ["workclass", "education", "marital-status", "occupation", "relationship", "race", "sex", "native-country"]

for col in categorical_columns:
    print(f"Column: {col}")
    print(data[col].value_counts())
    print("\n")
```

```python
data['workclass'].fillna("Unknown", inplace=True)
data['occupation'].fillna(data['occupation'].mode()[0], inplace=True)
```

```python
data['age'].fillna(data['age'].median(), inplace=True)
```

```python
data.dropna(inplace=True)
```

```python
# Define the categorical features and the label column
categorical_features = ["workclass", "education", "marital-status", "occupation", "relationship", "race", "sex", "native-country"]
label_column = "salary"

# Process the data for training
X, y, encoder, lb = process_data(
    data,
    categorical_features=categorical_features,
    label=label_column,
    training=True
)

```

```python
print("Processed Features Shape:", X.shape)
print("Processed Labels Shape:", y.shape)
print(X[:5])  # Show first 5 rows of processed features
print(y[:5])  # Show first 5 rows of processed labels
```
